# Q1. Write a python program to display all the header tags from wikipedia.org

```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.wikipedia.org/'
site = requests.get(url)
bs = BeautifulSoup(site.text)
header = ['h1','h2','h3','h4']
for i in bs.find_all(header):
    print('\n', i.name, '\t', i.text.strip())
```

# Q2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
link = 'https://www.imdb.com/chart/top'
site = requests.get(link)
bs = BeautifulSoup(site.content)
name = []
b=0
for i in bs.find_all('td',class_='titleColumn'): 
    if(b<100):
        name.append(i.text.split('\n')[2].strip())
        b=b+1
    else:
        break
year = []
b=0
for i in bs.find_all('span',class_='secondaryInfo'):
    if(b<100):
        year.append(i.text)
        b+=1
    else:
        break
rating = []
b=0
for i in bs.find_all('td',class_='ratingColumn imdbRating'):
    if(b<100):
        children=i.find('strong')
        rating.append(children.text)
        b+=1
    else:
        break
print(len(name),len(year),len(rating))
import pandas as pd
df = pd.DataFrame({'Name':name,'Year':year,'Rating':rating})
df
```

# Q4. Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp.

```python
url = 'https://meesho.com/bagsladies/pl/p7vbp'
site = requests.get(url)
bs = BeautifulSoup(site.content)
name = []
for i in bs.find_all('p',class_='Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS'):
    name.append(i.text)
price = []
for i in bs.find_all('div',class_='Card__BaseCard-sc-b3n78k-0 iLPHgK NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr'):
    price.append(i.text.split('₹')[1])
discount = []
for i in bs.find_all('p',class_='Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm'):
    discount.append(i.text.split(' ')[0].replace('₹',''))
df = pd.DataFrame({'Title':name,'Price':price,'Discount':discount})
df
```

# Q7. Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
url = 'https://coreyms.com/'
site = requests.get(url)
bs = BeautifulSoup(site.content)

heading = []
for i in bs.find_all('a',class_='entry-title-link'):
    heading.append(i.text)
print('\n Headings: \n',heading)

time = []
for i in bs.find_all('time',class_='entry-time'):
    time.append(i.text)
print('\n Time: \n',time)

content = []
for i in bs.find_all('div',class_='entry-content'):
    content.append(i.text.strip())
print('\n Content: \n',content)

video = []
for i in bs.find_all('iframe'):
    response = requests.get(i.get('src'))
    bs2 = BeautifulSoup(response.content)
    b=1
    for j in bs2.find_all('link'):
        if(b==2):
            k = j.get('href')
            video.append(k)
        b+=1
print('\n Video: \n',video)
```

# Q9. Write a python program to scrape details from dineout.co.in

```python
url = 'https://www.dineout.co.in/delhi-restaurants'
site = requests.get(url)
bs = BeautifulSoup(site.content)

## (i) Restaurant name
name = []
for i in bs.find_all('a',class_='restnt-name ellipsis'):
    name.append(i.text)
    
## (ii) Cuisine
cuisine = []
for i in bs.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(i.text.split('|')[1])
    
## (iii) Location
loc = []
for i in bs.find_all('div',class_='restnt-loc ellipsis'):
    loc.append(i.text.strip())

## (iv) Ratings
rating = []
for i in bs.find_all('div',class_='restnt-rating rating-4'):
    rating.append(i.text.strip())

## (v) Images URL
img = []
for i in bs.find_all('img',class_='no-img'):
    img.append(i['data-src'])
    
df = pd.DataFrame({'Name':name,'Cuisine':cuisine,'Location':loc,'Ratings':rating,'Image':img})
df
```

# Q10. Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
url = 'https://www.bewakoof.com/women-tshirts?ga_q=tshirts'
site = requests.get(url)
bs = BeautifulSoup(site.content)
name = []
price = []
img = []
b = 0
for i in bs.find_all('h3'):
    if(b<10):
        name.append(i.text)
        b+=1
    else:
        break
b = 0
for i in bs.find_all('span',class_='discountedPriceText'):
    if(b<10):
        price.append(i.text)
        b+=1
    else:
        break
b = 0
for i in bs.find_all('img',class_='productImgTag'):
    if(b<10):
        img.append(i['src'])
        b+=1
    else:
        break
df = pd.DataFrame({'Name':name,'Price':price,'Image':img})
df
```
