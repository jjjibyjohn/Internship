# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
# Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
# You need to find following details:
# A) Rank
# B) Name
# C) Artist
# D) Upload date
# E) Views

```python
# Calling the webdriver exe file
driver = webdriver.Chrome(r"C:\Users\jjjib\Desktop\Projects\FlipRobo\chromedriver.exe") 
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos' # giving the url to the variable
```

```python
driver.get(url) # Calling the url in the webdriver
```

```python
# Defining empty arrays for the data to be scraped
rank = []
name = []
art = []
date = []
view = []
```

```python
# Extracting all the rows
rows = driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr')
```

```python
# Extracting data from each row
for i in range(len(rows)):
    tag1 = rows[i].find_element_by_xpath("./td[1]") #Extracting rank column
    rank.append(tag1.text)
    
    tag2 = rows[i].find_element_by_xpath("./td[2]") #Extracting name column
    name.append(tag2.text.split('"')[1])
    
    tag3 = rows[i].find_element_by_xpath("./td[3]") #Extracting artist column
    art.append(tag3.text)
    
    tag4 = rows[i].find_element_by_xpath("./td[5]") #Extracting date column
    date.append(tag4.text)
    
    tag5 = rows[i].find_element_by_xpath("./td[4]") #Extracting views column
    view.append(tag5.text)
```

```python
#Creating a Dataframe for the extracted data
df = pd.DataFrame({'Rank':rank,
                   'Name':name,
                   'Artist':art,
                   'Date Uploaded':date,
                   'Views (billions)':view})
```

```python
df #Displaying the dataframe created
```

```python
# Defining empty arrays for the data to be scraped
nme = []
tpe = []
tsk = []
att = []
nin = []
nat = []
yr = []
```

```python
tag = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr') #Scraping all the rows

#Running the rows one at a time to scrape the required data
for j in tag[1:]:
    try:
        k = j.find_element_by_xpath('./td[1]/table/tbody/tr/td[2]/p/b/a') #Scraping the name
        nme.append(k.text)
    except:
        nme.append("-")
        
    try:
        k = j.find_element_by_xpath('./td[2]/p') #Scraping the data type
        tpe.append(k.text)
    except:
        tpe.append("-")
        
    try:
        k = j.find_element_by_xpath('./td[3]/p') #Scraping the task
        tsk.append(k.text)
    except:
        tsk.append("-")
        
    try:
        k = j.find_element_by_xpath('./td[4]/p') #Scraping the attribute
        att.append(k.text)
    except:
        att.append("-")
        
    try:
        k = j.find_element_by_xpath('./td[5]/p') #Scraping the no. of instances
        nin.append(k.text)
    except:
        nin.append("-")
    
    try:
        k = j.find_element_by_xpath('./td[6]/p') #Scraping the no. of attribute
        nat.append(k.text)
    except:
        nat.append("-")
        
    try:
        k = j.find_element_by_xpath('./td[7]/p') #Scraping the year
        yr.append(k.text)
    except:
        yr.append("-")
```

```python
#Creating a Dataframe for the extracted data
df = pd.DataFrame({'Dataset Name':nme,
                   'Data Type':tpe,
                   'Task':tsk,
                   'Attribute Type':att,
                   'No. of Instances':nin,
                   'No. of Attribute':nat, 
                   'Year':yr})
```
