# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Defining empty arrays for the data to be scraped
brand = []
name = []
prc = []
ret = []
deli = []
avail = []
purl = []
```

```python
# Forming a loop to scrape all the data required
for url in link:
    driver.get(url) #Calling each url one-by-one
    
    #Scraping brand name
    try:
        tag1 = driver.find_element_by_id('productTitle')
        brand.append(tag1.text.split()[0])
    except:
        brand.append("-")
    
    #Scraping Product Name
    try:
        tag2 = driver.find_element_by_id('productTitle')
        name.append(tag2.text)
    except:
        name.append("-")
        
    #Scraping Price
    try:
        tag3 = driver.find_element_by_xpath('//span[@class="a-price a-text-price a-size-medium apexPriceToPay"]/span[2]')
        prc.append(tag3.text)
    except:
        prc.append("-")
        
    #Scraping Return/Exchange
    try:
        tag4 = driver.find_element_by_id('RETURNS_POLICY')
        ret.append(tag4.text)
    except:
        ret.append("-")
        
    #Scraping Delivery 
    try:
        tag5 = driver.find_element_by_xpath('//div[@id="ddmDeliveryMessage"]/b[1]')
        deli.append(tag5.text)
    except:
        deli.append("-")
        
    #Scraping Availability
    try:
        tag6 = driver.find_element_by_xpath('//div[@id="availability"]/span[1]')
        avail.append(tag6.text)
    except:
        avail.append("-")
        
    #Scraping Product URL
    purl.append(url)
```

```python
#Creating a DataFrame of the scraped data
df = pd.DataFrame({"Brand Name":brand, 
                   "Name of the Product":name, 
                   "Price":prc,
                   "Return/Exchange":ret,
                   "Expected Delivery":deli,
                   "Availability":avail,
                   "Product URL":purl})
```

```python
#Creating an empty array to store the link for the products
link = []
```

```python
# Defining empty arrays for the data to be scraped
brand = []
name = []
clr = []
ram = []
rom = []
pcm = []
scm = []
dis = []
btt = []
prc = []
purl = []
```

```python
# Forming a loop to scrape all the data required
for url in link:
    driver.get(url) #Calling each url one-by-one
    
    #Scraping brand name
    try:
        tag1 = driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        brand.append(tag1.text.split()[0])
    except:
        brand.append("-")
    
    #Scraping Product Name
    try:
        tag2 = driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        name.append(tag2.text.split("(")[0])
    except:
        name.append("-")
        
    #Scraping Colour
    try:                                     
        tag3 = driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        clr.append(tag3.text.split("(")[1].split(",")[0])
    except:
        clr.append("-")
        
    #Scraping RAM
    try:
        tag4 = driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        ram.append(tag4.text.split("(")[-1].split(")")[0])
    except:
        ram.append("-")
    
    #Scraping ROM
    try:
        tag5 = driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        rom.append(tag5.text.split("(")[1].split(",")[1].split(")")[0])
    except:
        rom.append("-")
        
    #Scraping Primary Camera
    try:
        tag6 = driver.find_element_by_xpath('//div[@class="_2418kt"]/ul/li[3]')
        pcm.append(tag6.text.split("|")[0])
    except:
        pcm.append("-")
        
    #Scraping Secondary Camera
    try:
        tag7 = driver.find_element_by_xpath('//div[@class="_2418kt"]/ul/li[3]')
        scm.append(tag7.text.split("|")[1])
    except:
        scm.append("-")
        
    #Scraping Display
    try:
        tag8 = driver.find_element_by_xpath('//div[@class="_2418kt"]/ul/li[2]')
        dis.append(tag8.text)
    except:
        dis.append("-")
        
    #Scraping Battery
    try:
        tag9 = driver.find_element_by_xpath('//div[@class="_2418kt"]/ul/li[4]')
        btt.append(tag9.text)
    except:
        btt.append("-")
        
    #Scraping Price
    try:
        tag10 = driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        prc.append(tag10.text)
    except:
        prc.append("-")
        
    #Scraping Product URL
    purl.append(url)
```

```python
#Creating a DataFrame of the scraped data
df = pd.DataFrame({"Brand Name":brand, 
                   "Smartphone Name":name,
                   "Colour":clr,
                   "RAM":ram,
                   "Storage(ROM)":rom,
                   "Primary Camera":pcm,
                   "Secondary Camera":scm,
                   "Display Size":dis,
                   "Battery Capacity":btt,
                   "Price":prc,
                   "Product URL":purl})
```

# 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
#Calling the webdriver exe file
driver = webdriver.Chrome(r"C:\Users\jjjib\Desktop\Projects\FlipRobo\chromedriver.exe")
```

```python
driver.get("https://www.youtube.com") #Calling the website on the webdriver
```

```python
# Clicking on the first suggested video
btn = driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-browse/ytd-two-column-browse-results-renderer/div[1]/ytd-rich-grid-renderer/div[6]/ytd-rich-grid-row[1]/div/ytd-rich-item-renderer[1]/div/ytd-rich-grid-media/div[1]/div[2]/div[1]/h3/a')
btn.click()
```

```python
#Creating empty list to store the scraped data
cmt = []
up = []
tm = []
```

```python
j=k=l=0
while j<500:
    #Locating and scraping comments
    tag = driver.find_elements_by_xpath('//div[@class="style-scope ytd-expander"]/yt-formatted-string[2]')
    for i in tag:
        if j<500:
            cmt.append(i.text)
            j+=1
    
    #Locating and scraping upvotes
    tag = driver.find_elements_by_id('vote-count-middle')
    for i in tag:
        if k<500:
            if ((i.text)==''):  #Checking if the value is null
                up.append('-')
                k+=1
            else:
                up.append(i.text)
                k+=1

    #Locating and Scraping time of post for the comment
    tag = driver.find_elements_by_xpath('//yt-formatted-string[@class="published-time-text above-comment style-scope ytd-comment-renderer"]/a')
    for i in tag:
        if l<500:
            tm.append(i.text)
            l+=1
    
    load = driver.find_element_by_xpath('//div[@class="style-scope ytd-item-section-renderer"][4]')
    p = load.location_once_scrolled_into_view
```

```python
#Creating DataFrame of the scraped data
df = pd.DataFrame({'Comment':cmt,
                   'Upvote':up,
                   'Time when Posted':tm})
```

```python
#Calling the webdriver exe file
driver = webdriver.Chrome(r"C:\Users\jjjib\Desktop\Projects\FlipRobo\chromedriver.exe")
```

```python
driver.get("https://www.hostelworld.com/ ") #Calling the website on the webdriver
```

```python
#Locating and input into search bar
search = driver.find_element_by_id('search-input-field')
search.send_keys("London")
```

```python
#Selecting the first option from the drop down list
dd = driver.find_element_by_xpath('//ul[@id="predicted-search-results"]/li[2]')
dd.click()
```

```python
#Clicking the go button
go = driver.find_element_by_xpath('//button[@id="search-button"]')
go.click()
```

```python
#Defining empty array to store the scraped data
name = []
dis = []
rat = []
tot = []
ovr = []
pvt = []
dor = []
fac = []
des = []
i=0
```

```python
for i in range(0,3):
    
    tag1 = driver.find_elements_by_xpath('//h2[@class="title title-6"]/a')  #Locating Hostel name
    for j in tag1:
        name.append(j.text)
    
    tag2 = driver.find_elements_by_xpath('//span[@class="description"]')  #Locating Distance
    for j in tag2:
        dis.append(j.text.split("-")[1])
    
    tag3 = driver.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]/div[1]')  #Locating Rating
    for j in tag3:
        try:
            float(j.text)
            rat.append(j.text)
        except:
            rat.append("-")
    
    tag4 = driver.find_elements_by_xpath('//div[@class="reviews"]')  #Locating Total Reviews
    for j in tag4:
        try:
            tot.append(j.text)
        except:
            tot.append("-")
        
    tag5 = driver.find_elements_by_xpath('//div[@class="keyword"]/span')  #Locating Overall Reviews
    for j in tag5:
        if (i==0): 
            i+=1
        else:
            if (((j.text)=='No Rating') or ((j.text)=='No Rating')):
                ovr.append("-")
            else:
                ovr.append(j.text)
    
    tag6 = driver.find_elements_by_xpath('//div[@class="price-col"][1]')  #Locating Privates from Price
    for j in tag6:
        try:
            pvt.append(j.text.replace('\n',' '))
        except:
            pvt.append("-")
          
    tag7 = driver.find_elements_by_xpath('//div[@class="price-col"][2]')  #Locating Dorms from Price
    for j in tag7:
        try:
            dor.append(j.text.replace('\n',' '))
        except:
            dor.append("-")
    
    tag8 = driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')  #Locating Facilities
    for j in tag8:
        if (j.text==''):
            fac.append("-")
        else:
            fac.append(j.text.replace('\n',' '))
        
    tag9 = driver.find_elements_by_xpath('//div[@class="rating-factors prop-card-tablet rating-factors small"]')  #Locating Description
    for j in tag9:
        if (j.text==''):
            des.append("-")
        else:
            des.append(j.text.replace('\n',' '))
    
    driver.implicitly_wait(5)
    
    try:
        btn = driver.find_element_by_xpath('//div[@class="pagination-item pagination-next"]') #Clicking on next page option
        btn.click()
    except NoSuchElementException as e:
        pass
```

```python
print(len(name),len(dis),len(rat),len(tot),len(ovr),len(pvt),len(dor),len(fac),len(des))
```

```python
df = pd.DataFrame({'Name':name,
                   'Distance':dis,
                   'Rating':rat,
                   'Total Reviews':tot,
                   'Overall Review':ovr,
                   'Privates From':pvt,
                   'Dorms From':dor,
                   'Facilities':fac,
                   'Description':des})
```
